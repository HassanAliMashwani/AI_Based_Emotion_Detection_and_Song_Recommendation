```python
from datasets import load_dataset
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

```python
import os
os.environ["HF_DATASETS_CACHE"] = "/content/hf_cache"
```

```python
# ================================================
# 🟢 Load GoEmotions dataset (combined train + val)
# ================================================
ds = load_dataset("dair-ai/emotion")
train_ds = ds["train"]
val_ds = ds["validation"]

texts = list(train_ds["text"]) + list(val_ds["text"])
labels = [train_ds.features["label"].int2str(l) for l in train_ds["label"]] + \
         [val_ds.features["label"].int2str(l) for l in val_ds["label"]]

print(f"✅ Loaded {len(texts)} text samples and labels.")
```

```python
# ================================================
# 🟢 4) Map labels to your categories
# ================================================
# Map detailed emotion labels to simplified categories
label_map = {
    "joy": "happy",
    "sadness": "sad",
    "anger": "angry",
    "love": "happy",
    "fear": "sad",
    "surprise": "happy"
}

mapped_labels = [label_map.get(lbl, "neutral") for lbl in labels]
```

```python
# ================================================
# 🟢 5) Train text classification model
# ================================================
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(texts, mapped_labels)

print("✅ Model trained successfully!")
```

```python
# ================================================
# 🟢 6) Define song recommendations
# ================================================
songs = {
    "happy": [
        "Happy - Pharrell Williams",
        "Walking on Sunshine - Katrina & The Waves",
        "Can't Stop The Feeling! - Justin Timberlake",
        "Best Day Of My Life - American Authors"
    ],
    "sad": [
        "Someone Like You - Adele",
        "Fix You - Coldplay",
        "Hurt - Johnny Cash",
        "The Night We Met - Lord Huron"
    ],
    "angry": [
        "Break Stuff - Limp Bizkit",
        "Killing In The Name - Rage Against The Machine",
        "Duality - Slipknot",
        "Given Up - Linkin Park"
    ],
    "neutral": [
        "Let It Be - The Beatles",
        "Imagine - John Lennon",
        "Comfortably Numb - Pink Floyd",
        "Africa - Toto"
    ]
}
```

```python
# ================================================
# 🟢 7) Ask the user for input, predict mood, and recommend songs
# ================================================
msg = input("\n🗨️  How are you feeling? Describe your mood: ")

probs = model.predict_proba([msg])[0]
max_prob = max(probs)
predicted_mood = model.classes_[probs.argmax()]

if max_prob < 0.5:
    predicted_mood = "neutral"
print(f"\n🤖 I think you're feeling {predicted_mood.upper()}!")

print("\n🎵 Here are some songs for you:")
for song in songs[predicted_mood]:
    print(f"- {song}")
```
